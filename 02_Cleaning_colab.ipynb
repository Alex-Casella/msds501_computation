## what are common methods you know for filling in data?

- number: average/mean
-  deletion
-  informed guesses - domain knowledge
-  back fill / forward fill
-  model missingness

## more examples

- categorical data: "missing" (a very common and safe bet)
- categorical data: domain knowledge

- numerical data: 0 if numerical (can throw off distributions and be misleading if 0 has true value in the dataset)
- numerical data: -999 if numerical (can throw off distributions)

- Either: imputation (of mean, most common category, separate model for prediction of missing data etc.) but this destroys correlation

```python
# TODO: Fill missing values:
#   - CITY          -> "missing"        (safe categorical placeholder)
#   - UG_CLASS_YEAR -> 0               (distinct numeric sentinel)
#   - UG_SCHOOL     -> "not_undergrad" (uses domain knowledge)
# Hint: Series.fillna(value=..., inplace=True)
#
df["CITY"].fillna(value = "missing", inplace = True)
df["UG_CLASS_YEAR"].fillna(value = 0, inplace = True)
df["UG_SCHOOL"].fillna(value = "not_undergrad", inplace = True)

# TODO: Also drop rows in bev where Date == "#REF!".
# Hint: bev = bev[bev.Date != ...]

```

## when should you use one hot encoding?

- very, very low cardinality feature
- when using linear models such as linear regression or logistic regression
- in practice, one hot encoding is not often used
- comparing methods: https://scikit-learn.org/stable/auto_examples/preprocessing/plot_target_encoder.html#sphx-glr-auto-examples-preprocessing-plot-target-encoder-py

## when should you use label encoding?

- high cardinality feature
- when using non-linear models random forest
- in practice, label encoding is often used (because random forest and other tree-based models dominate data science)
